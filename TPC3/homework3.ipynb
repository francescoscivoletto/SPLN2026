```python
import spacy
from spacy.matcher import Matcher
from itertools import combinations
```

```python
nlp = spacy.load("pt_core_news_sm")
# Create the matcher object
matcher = Matcher(nlp.vocab)
```

```python
# Pattern: one or more proper nouns recognized as person entities
character_pattern = [
    {"ENT_TYPE": "PER", "POS": "PROPN", "OP": "+"}
]

matcher.add("CHARACTER", [character_pattern], greedy="LONGEST")
```

```python
def is_valid_name(name):
    if not name:
        return False

    if not name[0].isupper():
        return False

    invalid_tokens = ["\n", "©", '"', "www", ".org", "Ltda", "Limited"]

    for token in invalid_tokens:
        if token in name:
            return False

    return True
```

```python
with open("harrypotter.txt", "r", encoding="utf-8") as file:
    text = file.read()
```

```python
doc = nlp(text)
characters = set()
relationships = set()

# Analyze the text sentence by sentence
for sentence in doc.sents:
    sentence_doc = sentence.as_doc()
    matches = matcher(sentence_doc)

    characters_in_sentence = []

    for match_id, start, end in matches:
        name = sentence_doc[start:end].text.strip()

        if is_valid_name(name):
            characters_in_sentence.append(name)
            characters.add(name)

    unique_characters = sorted(set(characters_in_sentence))

    # Generate all possible pairs of characters in the same sentence
    for char1, char2 in combinations(unique_characters, 2):
        relationships.add((char1, char2))
```

```python
# Print the extracted characters
print("=== EXTRACTED CHARACTERS ===")
for character in sorted(characters):
    print(character)

# Print the extracted relationships
print("\n=== CHARACTER RELATIONSHIPS ===")
for char1, char2 in sorted(relationships):
    print(f"{char1} <-> {char2}")
```
